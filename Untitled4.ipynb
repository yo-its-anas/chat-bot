```python
# Step 2: Import the necessary libraries
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import torch

# Step 3: Load pre-trained FLAN-T5 model and tokenizer from Hugging Face
model_name = "google/flan-t5-large"  # You can use "small", "base", or "xl" depending on your needs
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Step 4: Interactive Q&A chatbot function
def chat_with_bot():
    print("Chatbot: Hello! I can answer your questions about Generative AI. Type 'bye' to end the conversation.\n")

    while True:
        # User input
        user_input = input("You: ")

        if user_input.lower() == "bye":
            print("Chatbot: Goodbye!")
            break

        # Format input as a question for the model
        input_text = f"Answer this question about Generative AI: {user_input}"

        # Tokenize the input text
        input_ids = tokenizer(input_text, return_tensors="pt").input_ids

        # Generate a response using the model
        output_ids = model.generate(input_ids, max_length=200)

        # Decode the model's response
        bot_response = tokenizer.decode(output_ids[0], skip_special_tokens=True)
        print(f"Chatbot: {bot_response}\n")

# Step 5: Run the chatbot
chat_with_bot()
```
